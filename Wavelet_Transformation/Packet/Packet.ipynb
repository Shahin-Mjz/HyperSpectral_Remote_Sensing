```python
import os
import cv2
import pywt
import numpy as np
from termcolor import colored
import matplotlib.pyplot as plt
from pywt._doc_utils import wavedec2_keys, draw_2d_wp_basis
```

```python
path_str = os.getcwd() + '/'
img_name_str = 'cameraman.jpg'
```

```python
def min_max_scaler(input_img: np.ndarray, data_range: tuple=(0, 1)) -> np.ndarray:
    img = input_img.copy()
    a, b = data_range[0], data_range[1]
    img = (img - img.min()) / (img.max() - img.min())
    img *= (b - a)
    img += a
    return img
```

```python
img_orig_ndarray = cv2.imread(path_str + img_name_str, 0)
img_orig_ndarray = min_max_scaler(input_img=img_orig_ndarray, data_range=(0, 1))
img_orig_ndarray = img_orig_ndarray.astype(np.float32)
```

```python
print(f'\nShape of the Original Image: {colored(img_orig_ndarray.shape, "blue")}\n')
print(f' ... and its {colored("Data Type", "blue")}: {colored(img_orig_ndarray.dtype, "blue")}\n')
```

```python
plt.figure(1)
plt.imshow(img_orig_ndarray, cmap='gray'), plt.axis('off'), plt.title('Original Image', fontsize=10)
plt.show()
```
