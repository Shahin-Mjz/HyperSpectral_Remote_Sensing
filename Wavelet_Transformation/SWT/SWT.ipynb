### <span style='color:gray'> $\bf{\text{Implementation:}}$ </span>

```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
from skimage.util import random_noise
from skimage.restoration import denoise_wavelet
```

```python
# Global Variables:
orig_img_ndarr = pywt.data.camera()

plt.figure()
plt.subplot(), plt.imshow(orig_img_ndarr, cmap='gray'), plt.axis('off'), plt.title('CameraMan')
plt.show()
```

```python
wavelet_list = pywt.wavelist()
print(f'\nWavelets:\n\n{wavelet_list}')
```

- <span style='color:red'>$\bf{\text{Stationary Wavelet Transform (SWT)}}$</span>


- - [x] <span style='color: blue'>$\text{Multilevel 1D swt:}$</span>


- - - $\text{Check the}$ $\begin{cases}
\text{axis = 0} \\
\text{axis = 1}
\end{cases}$


- - - $\text{Check the}$ $\begin{cases}
\text{level = 2} \\
\text{level = 3}
\end{cases}$

```python
levels = 3
tr_img_list1 = pywt.swt(data=orig_img_ndarr, wavelet='sym2', level=levels, axis=0)

index = 0
plt.figure(figsize=(14, 14))
for level in range(1, levels + 1):
    index += 1
    plt.subplot(levels, 2, index), plt.imshow(tr_img_list1[levels - level][0], cmap='gray'), plt.axis('off')
    plt.title(f'App. for Level {level}', fontsize=10)
    index += 1
    plt.subplot(levels, 2, index), plt.imshow(tr_img_list1[levels - level][1], cmap='gray'), plt.axis('off')
    plt.title(f'Det. for Level {level}', fontsize=10)
plt.show()
```

- <span style='color:red'>$\bf{\text{Stationary Wavelet Transform (SWT)}}$</span>


- - [x] <span style='color: blue'>$\text{Multilevel 2D swt2:}$</span>


- - - $\text{Check the}$ $\begin{cases}
\text{axes = (0, 1)} \\
\text{axes = (1, 0)}
\end{cases}$


- - - $\text{Check the}$ $\begin{cases}
\text{level = 1} \\
\text{level = 2}
\end{cases}$


><span style='color:green'>$\bf{\text{Note:}}$</span>
$\text{Axes must be unique.}$

```python
levels = 3
tr_img_list2 = pywt.swt2(data=orig_img_ndarr, wavelet='sym2', level=levels, axes=(0, 1))

index = 0
plt.figure(figsize=(12, 12))
for level in range(1, levels + 1):
    index += 1
    plt.subplot(levels, 4, index), plt.imshow(tr_img_list2[levels - level][0], cmap='gray'), plt.axis('off')
    plt.title(f'App. for Level {level}', fontsize=9)
    index += 1
    plt.subplot(levels, 4, index), plt.imshow(tr_img_list2[levels - level][1][0], cmap='gray'), plt.axis('off')
    plt.title(f'Det. Horizontal for Level {level}', fontsize=9)
    index += 1
    plt.subplot(levels, 4, index), plt.imshow(tr_img_list2[levels - level][1][1], cmap='gray'), plt.axis('off')
    plt.title(f'Det. Vertical for Level {level}', fontsize=9)
    index += 1
    plt.subplot(levels, 4, index), plt.imshow(tr_img_list2[levels - level][1][2], cmap='gray'), plt.axis('off')
    plt.title(f'Det. Diagonal for Level {level}', fontsize=9)
plt.show()
```

- <span style='color:red'>$\bf{\text{Inverse Stationary Wavelet Transform (ISWT)}}$</span>


- - [x] <span style='color:blue'>$\text{Multilevel 1D iswt:}$</span>

```python
inv_img_ndarr1 = pywt.iswt(tr_img_list1, 'sym2')
plt.figure(figsize=(12, 12))
plt.subplot(1, 2, 1), plt.imshow(orig_img_ndarr, cmap='gray'), plt.axis('off'), \
plt.title('Original Image', fontsize=10)
plt.subplot(1, 2, 2), plt.imshow(inv_img_ndarr1, cmap='gray'), plt.axis('off'), \
plt.title('ISWT (Original Image)', fontsize=10)
plt.show()
```

- <span style='color:red'>$\bf{\text{Inverse Stationary Wavelet Transform (ISWT)}}$</span>


- - [x] <span style='color:blue'>$\text{Multilevel 2D iswt2:}$</span>

```python
inv_img_ndarr2 = pywt.iswt2(tr_img_list2, 'sym2')
plt.figure(figsize=(12, 12))
plt.subplot(1, 2, 1), plt.imshow(orig_img_ndarr, cmap='gray'), plt.axis('off'), \
plt.title('Original Image', fontsize=10)
plt.subplot(1, 2, 2), plt.imshow(inv_img_ndarr2, cmap='gray'), plt.axis('off'), \
plt.title('ISWT (Original Image)', fontsize=10)
plt.show()
```

- <span style='color:red'>$\bf{\text{Comparison between DWT & SWT}}$</span>


- - [x] <span style='color:blue'>$\text{Conclusion}$</span>

$$\text{Denoising by using SWT}$$

<img src='denoising.png' style='width:900px; height:900px'>

```python
m = orig_img_ndarr.size
MSE1 = np.unique(np.sum((np.abs(orig_img_ndarr - inv_img_ndarr1)**2))/m)
print(f'Mean Squared Error for 1D SWT (MSE): {MSE1[0]}')
```

```python
m = orig_img_ndarr.size
MSE2 = np.unique(np.sum((np.abs(orig_img_ndarr - inv_img_ndarr2)**2))/m)
print(f'Mean Squared Error for 2D SWT(MSE): {MSE2[0]}')
```
